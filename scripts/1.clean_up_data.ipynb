```python
import pandas as pd
```

```python
csv_file = 'first_data'
csv_file_name = csv_file + '.csv'
csv_file_path = os.path.join('csv', csv_file_name)

img_wrong_path = os.path.join('img', 'wrong')

print(f'File csv - {csv_file_name}, path - {csv_file_path}')
print(f'Path wrong img - {img_wrong_path}')
```

```python
df = pd.read_csv(csv_file_path, header=0)
old_rows_count = df.shape[0]

print(f'Records - {df.shape[0]}')

df.head(3)
```

# Удаление плохих изображений

## Получение списка всех плохих изображений в базе

```python
wrong_list = os.listdir(img_wrong_path)
```

```python
img_list = df['img'].tolist()
```

```python
wrong_list = [x for x in wrong_list if x in img_list]
```

```python
print(wrong_list)
```

```python
print(len(wrong_list))
```

## Получение индексов плохих изображений

```python
def get_index(name_wrong_img):
  return df[df['img'] == name_wrong_img].index.tolist()[0]
```

```python
wrong_list_index = [get_index(name_wrong_img) for name_wrong_img in wrong_list]
```

```python
print(wrong_list_index)
```

```python
print(len(wrong_list_index))
```

## Удаление плохих изображений из базы

```python
df = df.drop(df.index[wrong_list_index])
```

```python
df.shape[0]
```

```python
assert(df.shape[0] + len(wrong_list) == old_rows_count)
```

```python
df.to_csv(csv_file_path, index=False)
```
